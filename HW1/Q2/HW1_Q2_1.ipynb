```python
import numpy as np
import matplotlib.pyplot as plt

# Load the data
data = np.load('/content/data.npy', allow_pickle=True)  # Change to the actual path of your data file

# Split the data into training and testing sets
train_size = int(0.8 * len(data))  # 80% for training, 20% for testing
train_data = data[:train_size]
test_data = data[train_size:]

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(range(len(train_data)), train_data, label="Training Data", color="blue")
plt.plot(range(len(train_data), len(data)), test_data, label="Testing Data", color="orange")
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Training and Testing Data")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# بارگذاری داده‌ها
data = np.load('/content/data.npy', allow_pickle=True)

# ایجاد ویژگی‌ها و هدف
X = np.arange(len(data))  # ویژگی‌ها
y = data  # مقادیر هدف

# محاسبه میانگین x و y
x_mean = np.mean(X)
y_mean = np.mean(y)

# محاسبه شیب (w1)
numerator = np.sum((X - x_mean) * (y - y_mean))  # صورت
denominator = np.sum((X - x_mean)**2)  # مخرج
w1 = numerator / denominator

# محاسبه عرض از مبدأ (w0)
w0 = y_mean - w1 * x_mean

# پیش‌بینی مقادیر
y_pred = w1 * X + w0

# محاسبه MSE
mse = np.mean((y - y_pred)**2)
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Model: y = {w1:.4f}x + {w0:.4f}")

# رسم نمودار
plt.figure(figsize=(10, 6))
plt.scatter(X, y, label="Actual Data", color="blue", alpha=0.6)
plt.plot(X, y_pred, label="Linear Fit (Degree 1)", color="red", linewidth=2)
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Linear Regression (Degree 1) Fit")
plt.legend()
plt.grid(True)
plt.show()
```

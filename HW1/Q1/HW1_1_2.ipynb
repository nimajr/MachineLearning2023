```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('/content/BankChurners.csv')
data
```

```python
# انتخاب چند ویژگی
selected_features = ['Customer_Age', 'Credit_Limit', 'Total_Trans_Amt', 'Avg_Utilization_Ratio']
```

```python
# رسم نمودار پراکندگی
sns.pairplot(data[selected_features])
plt.show()
```

```python
sns.histplot(data['Customer_Age'], kde=True)
plt.title("Distribution of Customer Age")
plt.show()
```

```python
sns.kdeplot(data['Credit_Limit'], shade=True)
plt.title("Distribution of Credit Limit")
plt.show()
```

```python
sns.jointplot(data=data, x='Customer_Age', y='Credit_Limit', kind='kde')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('/content/BankChurners.csv')
data
```

```python
# رسم نمودار میله‌ای برای یکی از ویژگی‌های طبقه‌بندی‌شده
sns.countplot(data=data, x='Gender')
plt.title("Distribution of Gender")
plt.show()

sns.countplot(data=data, x='Education_Level')
plt.title("Distribution of Education Level")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# تبدیل ویژگی‌های طبقه‌بندی‌شده به عددی
encoder = LabelEncoder()
data['Gender_Encoded'] = encoder.fit_transform(data['Gender'])  # Male = 1, Female = 0
data['Education_Level_Encoded'] = encoder.fit_transform(data['Education_Level'])
```

```python
# انتخاب ویژگی‌های عددی مرتبط
selected_features = ['Gender_Encoded', 'Education_Level_Encoded', 'Customer_Age', 'Credit_Limit', 'Avg_Utilization_Ratio']

# محاسبه ماتریس همبستگی
correlation_matrix = data[selected_features].corr()
```

```python

# رسم نقشه حرارتی
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap for Selected Features")
plt.show()
```

```python
sns.scatterplot(data=data, x='Credit_Limit', y='Avg_Utilization_Ratio')
plt.title("Scatter Plot of Credit Limit vs Avg Utilization Ratio")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def mc_culloch_pitts_triangle(x, y):
    """
    Simulates the McCulloch-Pitts neural network for detecting points inside a triangle.
    """
    # Conditions for being inside the triangle
    condition1 = y >= 0  # Above line BC
    condition2 = y <= -2 * x + 6  # Below line AB
    condition3 = y <= 2 * x - 2  # Below line CA

    # Logical AND of all conditions
    return int(condition1 and condition2 and condition3)

# Generate a grid of points for visualization
x_coords = np.linspace(0, 4, 200)
y_coords = np.linspace(0, 3, 200)
x_grid, y_grid = np.meshgrid(x_coords, y_coords)

# Apply the neural network to each point
output = np.array([[mc_culloch_pitts_triangle(x, y) for x, y in zip(row_x, row_y)]
                   for row_x, row_y in zip(x_grid, y_grid)])

# Visualize the results
plt.figure(figsize=(8, 6))
plt.contourf(x_grid, y_grid, output, levels=[-0.5, 0.5, 1.5], colors=['red', 'green'], alpha=0.6)
plt.plot([1, 3], [0, 0], 'b-', label='BC')  # Line BC
plt.plot([2, 3], [2, 0], 'b-', label='AB')  # Line AB
plt.plot([1, 2], [0, 2], 'b-', label='CA')  # Line CA
plt.title("McCulloch-Pitts Neural Network for Triangle Detection")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def mc_culloch_pitts_triangle(x, y):
    """
    Simulates the McCulloch-Pitts neural network for detecting points inside a triangle.
    """
    # Conditions for being inside the triangle
    condition1 = y >= 0  # Above line BC
    condition2 = y <= -2 * x + 6  # Below line AB
    condition3 = y <= 2 * x - 2  # Below line CA

    # Logical AND of all conditions
    return int(condition1 and condition2 and condition3)

# Generate random points
np.random.seed(42)  # For reproducibility
num_points = 2000
x_random = np.random.uniform(0, 4, num_points)  # X in range [0, 4]
y_random = np.random.uniform(0, 3, num_points)  # Y in range [0, 3]

# Apply the neural network to each random point
outputs = np.array([mc_culloch_pitts_triangle(x, y) for x, y in zip(x_random, y_random)])

# Separate points based on output
inside_x = x_random[outputs == 1]
inside_y = y_random[outputs == 1]
outside_x = x_random[outputs == 0]
outside_y = y_random[outputs == 0]

# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(inside_x, inside_y, color='green', label='Inside Triangle (Output = 1)', alpha=0.6)
plt.scatter(outside_x, outside_y, color='red', label='Outside Triangle (Output = 0)', alpha=0.6)
plt.plot([1, 3], [0, 0], 'b-', label='BC')  # Line BC
plt.plot([2, 3], [2, 0], 'b-', label='AB')  # Line AB
plt.plot([1, 2], [0, 2], 'b-', label='CA')  # Line CA
plt.title("Random Points and Neural Network Classification")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```
